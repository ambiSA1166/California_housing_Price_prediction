## spliting

```python
## independent and dependent features
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
print(X.head())
print("--"*10)
print(y.head())
```

```python
#Train-Test-split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3 , random_state = 42)
```

```python
print("X_train shape:",X_train.shape)
print("y_train shape:",y_train.shape)
print("X_test shape:",X_test.shape)
print("y_test shape:",y_test.shape)
```

## standard scaling (standardization)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(X_train,y_train)
```

```python
print(model_lr.coef_) #gives you the learned coefficients (weights) of a Linear Regression model after training
print(model_lr.intercept_) #The intercept is the value your model predicts when all features are 0.
```

```python
# the parameters the model has been trained
model_lr.get_params()
```

```python
## prediction with X_test
lr_pred = model_lr.predict(X_test)
```

```python
lr_pred
```

```python
# plotting a scatter plot for the prediction
plt.scatter(y_test,lr_pred)
plt.show()
```

```python
residual = y_test - lr_pred #error
```

```python
#plotting the residuals
sns.displot(residual , kind = "kde")
plt.show()
```

```python
# scatter plot wrt residual ana predictions
plt.scatter(lr_pred , residual)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print("MAE:", mean_absolute_error(y_test, lr_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, lr_pred)))
```

Formula

#### R^2 = 1 - SSR/SST

R^2 = coefficient of determination SSR = sum of squares of residuals SST = total sum of squares

```python
score = r2_score(y_test, lr_pred)
print("R2:", score)
```

#### Adjusted R2 = 1 – [(1-R2)*(n-1)/(n-k-1)]

where:

R2: The R2 of the model n: The number of observations k: The number of predictor variables

```python
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
